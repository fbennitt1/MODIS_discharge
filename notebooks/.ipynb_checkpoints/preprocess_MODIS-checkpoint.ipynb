```python
data_path = '/nas/cee-water/cjgleason/data/MODIS/bands_1_2_masked_2025_06_25/'
```

```python
# file_path = os.path.join(data_path, '1749100_GRDC_MYD.nc')
# file_paths = [os.path.join(data_path, '1291100_GRDC_MOD.nc'), os.path.join(data_path, '1291100_GRDC_MYD.nc')]
file_paths = [os.path.join(data_path, '07263450_USGS_MOD_2000_2021_1D.nc'), os.path.join(data_path, '07263450_USGS_MYD_2000_2021_1D.nc')]
```

```python
da_merged = xr.Dataset()
# file_paths_list =[da1,da2]

for file in file_paths:

        da_merged = xr.merge([da_merged, xr.open_dataset(file, engine='h5netcdf',)],
                             compat='override')
```

```python
## Reshape for plotting
# Extract longitude
llon = da_merged.coords['lon'].values
llat = da_merged.coords['latitude'].values

# Setup for reshape
test = llon[0]

for i, value in enumerate(llon):
    if i != 0 and value == test:
        lon_dim = i
        break
                              
n_pixels = len(llon)

lat_dim = int(n_pixels / lon_dim)

assert lon_dim * lat_dim == n_pixels, "Grid shape mismatch"   
```

```python
da2d = da_merged['surf2'].values.reshape(lat_dim, lon_dim, -1)  # if there's also 'date'
dims = ('latitude', 'lon', 'date')  # or without 'date' if time isn't included
coords = {'date': da_merged['date']}

# now build a new DataArray
da2 = xr.DataArray(da2d, dims=dims, coords=coords)
```

```python
# lat2d, lon2d = np.meshgrid(lat_dim, lon_dim, indexing='ij')

lat2d = np.reshape(llat, (lat_dim, lon_dim))
lon2d = np.reshape(llon, (lat_dim, lon_dim))
```

```python
da_merged = da_merged.assign_coords({
    'lat2d': ('pixel', lat2d),
    'lon2d': ('pixel', lon2d),
})
```
