```python
import os
# import time

import h5netcdf
import matplotlib.pyplot as plt
# import math
import numpy as np
import pandas as pd
import xarray as xr
```

### Read-in

```python
data_path = '/nas/cee-water/cjgleason/data/MODIS/bands_1_2_masked_2025_06_25/'
```

```python
file_path = os.path.join(data_path, '1749100_GRDC_MYD.nc')
file_paths = [os.path.join(data_path, '1749100_GRDC_MOD.nc'), os.path.join(data_path, '1749100_GRDC_MYD.nc')]
```

```python
ds = xr.open_dataset(file_paths[0], engine='h5netcdf')
# ds = xr.open_mfdataset(file_paths, engine='h5netcdf')
```

```python
ds
```

```python
# Calculate the standard deviation along the 'y' dimension
std_y = ds.std(dim='date')

# Calculate the mean along the 'y' dimension
mean_y = ds.mean(dim='date')

# Calculate the coefficient of variation along the 'y' dimension
cv_y = std_y / mean_y

# You can also calculate the CV for multiple dimensions
# cv_xy = da.std(dim=['x', 'y']) / da.mean(dim=['x', 'y'])
```

```python
cv_y["surf2"].plot(
    x="lon", y="lat",
    cmap="cubehelix",
    figsize=(8, 6), robust=True
)
```
